```python
def finddate(expression):
        
    year=expression.split('-')[0]
    month=expression.split('-')[1]
    date=int(year+month)

    return date
```

```python
df=df0
```

```python
df.drop('Subject',axis=1,inplace=True)
df.drop('Country',axis=1,inplace=True)
df.drop('Unit',axis=1,inplace=True)
df.drop('Unit Code',axis=1,inplace=True)
df.drop('PowerCode Code',axis=1,inplace=True)
df.drop('PowerCode',axis=1,inplace=True)
df.drop('Reference Period Code',axis=1,inplace=True)
df.drop('FREQUENCY',axis=1,inplace=True)
df.drop('Frequency',axis=1,inplace=True)
df.drop('Reference Period',axis=1,inplace=True)
df.drop('Flag Codes',axis=1,inplace=True)
df.drop('Time',axis=1,inplace=True)
df.drop('Flags',axis=1,inplace=True)
```

```python
#Turkey_Indices
```

```python
Turkey_BCI=df[(
    (df['LOCATION']=='TUR')   
   &(df['SUBJECT']=='BSCICP03') 
    )]
Turkey_BCI=Turkey_BCI.rename(columns = {'Value':'BCI_TUR'})
Turkey_CCI=df[(
    (df['LOCATION']=='TUR')   
   &(df['SUBJECT']=='CSCICP03') 
    )]
Turkey_CCI=Turkey_CCI.rename(columns = {'Value':'CCI_TUR'})
Turkey_CLI=df[(
    (df['LOCATION']=='TUR')   
   &(df['SUBJECT']=='LOLITONO') 
    )]
Turkey_CLI=Turkey_CLI.rename(columns = {'Value':'CLI_TUR'})
Turkey_GDP=df[(
    (df['LOCATION']=='TUR')   
   &(df['SUBJECT']=='LORSGPNO') 
    )]
Turkey_GDP=Turkey_GDP.rename(columns = {'Value':'GDP_TUR'})

# BCI 'BSCICP03'=>'OECD Standardised BCI, Amplitude adjusted (Long term average=100), sa',
# CCI 'CSCICP03'=>'OECD Standardised CCI, Amplitude adjusted (Long term average=100), sa',
# CLI 'LOLITONO'=>'Normalised (CLI)'
# GDP 'LORSGPNO'=>'Normalised (GDP)'
```

```python
Turkey_BCI['BCI_TUR-1'] = Turkey_BCI['BCI_TUR'].shift(1)
Turkey_BCI['BCI_TUR-2'] = Turkey_BCI['BCI_TUR'].shift(2)
Turkey_BCI['BCI_TUR-3'] = Turkey_BCI['BCI_TUR'].shift(3)
Turkey_BCI['BCI_TUR-6'] = Turkey_BCI['BCI_TUR'].shift(6)
Turkey_BCI['BCI_TUR-12'] = Turkey_BCI['BCI_TUR'].shift(12)
```

```python
Turkey_CCI['CCI_TUR-1'] = Turkey_CCI['CCI_TUR'].shift(1)
Turkey_CCI['CCI_TUR-2'] = Turkey_CCI['CCI_TUR'].shift(2)
Turkey_CCI['CCI_TUR-3'] = Turkey_CCI['CCI_TUR'].shift(3)
Turkey_CCI['CCI_TUR-6'] = Turkey_CCI['CCI_TUR'].shift(6)
Turkey_CCI['CCI_TUR-12'] = Turkey_CCI['CCI_TUR'].shift(12)
```

```python
Turkey_CLI['CLI_TUR-1'] = Turkey_CLI['CLI_TUR'].shift(1)
Turkey_CLI['CLI_TUR-2'] = Turkey_CLI['CLI_TUR'].shift(2)
Turkey_CLI['CLI_TUR-3'] = Turkey_CLI['CLI_TUR'].shift(3)
Turkey_CLI['CLI_TUR-6'] = Turkey_CLI['CLI_TUR'].shift(6)
Turkey_CLI['CLI_TUR-12'] = Turkey_CLI['CLI_TUR'].shift(12)
```

```python
Turkey_GDP['GDP_TUR-1'] = Turkey_GDP['GDP_TUR'].shift(1)
Turkey_GDP['GDP_TUR-2'] = Turkey_GDP['GDP_TUR'].shift(2)
Turkey_GDP['GDP_TUR-3'] = Turkey_GDP['GDP_TUR'].shift(3)
Turkey_GDP['GDP_TUR-6'] = Turkey_GDP['GDP_TUR'].shift(6)
Turkey_GDP['GDP_TUR-12'] = Turkey_GDP['GDP_TUR'].shift(12)
```

```python
mergedCSV = Turkey_BCI[['TIME','BCI_TUR','BCI_TUR-1','BCI_TUR-2',
                                     'BCI_TUR-3','BCI_TUR-6','BCI_TUR-12']].merge(Turkey_CCI[['TIME','CCI_TUR','CCI_TUR-1','CCI_TUR-2',
                                    'CCI_TUR-3','CCI_TUR-6','CCI_TUR-12']], 
            on = 'TIME',how = 'left').merge(Turkey_CLI[['TIME','CLI_TUR','CLI_TUR-1','CLI_TUR-2',
                                    'CLI_TUR-3','CLI_TUR-6','CLI_TUR-12']], 
            on = 'TIME',how = 'left').merge(Turkey_GDP[['TIME','GDP_TUR','GDP_TUR-1','GDP_TUR-2',
                                    'GDP_TUR-3','GDP_TUR-6','GDP_TUR-12']], 
            on = 'TIME',how = 'left')

# Convert TIME column to Date and drop N/A
mergedCSV=mergedCSV.rename(columns = {'TIME':'Date'})

for num in range(0,mergedCSV.shape[0]):
    mergedCSV.iloc[num,0]=finddate(mergedCSV.iloc[num,0])
mergedCSV=mergedCSV.dropna()

#save resulting output file
mergedCSV.to_csv('BCI_CCI_CLI_GDP_TUR.csv',index = False)
```
