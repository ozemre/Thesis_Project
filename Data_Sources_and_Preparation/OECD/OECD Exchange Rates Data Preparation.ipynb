```python
import numpy as np
import pandas as pd
```

```python
df0 = pd.read_csv('OECD_Data\ExchangeRates.csv')
```

```python
df0
```

```python
def finddate(expression):
        
    year=expression.split('-')[0]
    month=expression.split('-')[1]
    date=int(year+month)

    return date
```

```python
df=df0
```

```python
df.drop('SUBJECT',axis=1,inplace=True)
df.drop('Subject',axis=1,inplace=True)
df.drop('Country',axis=1,inplace=True)
df.drop('MEASURE',axis=1,inplace=True)
df.drop('Measure',axis=1,inplace=True)
df.drop('Unit Code',axis=1,inplace=True)
df.drop('Unit',axis=1,inplace=True)
df.drop('PowerCode Code',axis=1,inplace=True)
df.drop('PowerCode',axis=1,inplace=True)
df.drop('Reference Period Code',axis=1,inplace=True)
df.drop('FREQUENCY',axis=1,inplace=True)
df.drop('Frequency',axis=1,inplace=True)
df.drop('Reference Period',axis=1,inplace=True)
df.drop('Flag Codes',axis=1,inplace=True)
df.drop('Time',axis=1,inplace=True)
df.drop('Flags',axis=1,inplace=True)
```

```python
CHN_EXH=df[(
    (df['LOCATION']=='CHN') 
       )]
```

```python
TUR_EXH=df[(
    (df['LOCATION']=='TUR') 
    )]
```

```python
UK_EXH=df[(
    (df['LOCATION']=='GBR') 
    )]
```

```python
TUR_EXH=TUR_EXH.rename(columns = {'Value':'TRY'})
TUR_EXH=TUR_EXH.rename(columns = {'TIME':'Date'})
```

```python
TUR_EXH['TRY-1'] = TUR_EXH['TRY'].shift(1)
TUR_EXH['TRY-2'] = TUR_EXH['TRY'].shift(2)
TUR_EXH['TRY-3'] = TUR_EXH['TRY'].shift(3)
TUR_EXH['TRY-6'] = TUR_EXH['TRY'].shift(6)
TUR_EXH['TRY-12'] = TUR_EXH['TRY'].shift(12)
```

```python
TUR_EXH
```

```python
for num in range(0,TUR_EXH.shape[0]):
    TUR_EXH.iloc[num,1]=finddate(TUR_EXH.iloc[num,1])
```

```python
CHN_EXH=CHN_EXH.rename(columns = {'Value':'CNY'})
CHN_EXH=CHN_EXH.rename(columns = {'TIME':'Date'})
```

```python
CHN_EXH['CNY-1'] = CHN_EXH['CNY'].shift(1)
CHN_EXH['CNY-2'] = CHN_EXH['CNY'].shift(2)
CHN_EXH['CNY-3'] = CHN_EXH['CNY'].shift(3)
CHN_EXH['CNY-6'] = CHN_EXH['CNY'].shift(6)
CHN_EXH['CNY-12'] = CHN_EXH['CNY'].shift(12)
```

```python
for num in range(0,CHN_EXH.shape[0]):
    CHN_EXH.iloc[num,1]=finddate(CHN_EXH.iloc[num,1])
```

```python
UK_EXH=UK_EXH.rename(columns = {'Value':'GBP'})
UK_EXH=UK_EXH.rename(columns = {'TIME':'Date'})
```

```python
UK_EXH['GBP-1'] = UK_EXH['GBP'].shift(1)
UK_EXH['GBP-2'] = UK_EXH['GBP'].shift(2)
UK_EXH['GBP-3'] = UK_EXH['GBP'].shift(3)
UK_EXH['GBP-6'] = UK_EXH['GBP'].shift(6)
UK_EXH['GBP-12'] = UK_EXH['GBP'].shift(12)
```

```python
for num in range(0,UK_EXH.shape[0]):
    UK_EXH.iloc[num,1]=finddate(UK_EXH.iloc[num,1])
```

```python
Turkey_UK_EXH = TUR_EXH[['Date','TRY','TRY-1','TRY-2',
                                     'TRY-3','TRY-6','TRY-12']].merge(UK_EXH[['Date','GBP','GBP-1','GBP-2',
                                    'GBP-3','GBP-6','GBP-12']], 
            on = 'Date',how = 'left')

# Reindex main features for the first columns 'TRY' , 'GBP'

Turkey_UK_EXH=Turkey_UK_EXH.reindex(columns=[ 'TRY','GBP', 'Date','TRY-1', 'TRY-2', 'TRY-3', 'TRY-6', 'TRY-12', 
       'GBP-1', 'GBP-2', 'GBP-3', 'GBP-6', 'GBP-12'])

#save resulting output file

```

```python
China_UK_EXH = CHN_EXH[['Date','CNY','CNY-1','CNY-2',
                                     'CNY-3','CNY-6','CNY-12']].merge(UK_EXH[['Date','GBP','GBP-1','GBP-2',
                                    'GBP-3','GBP-6','GBP-12']], 
            on = 'Date',how = 'left')


# Reindex main features for the first columns 'CNY' , 'GBP'

China_UK_EXH=China_UK_EXH.reindex(columns=[ 'CNY', 'GBP', 'Date', 'CNY-1', 'CNY-2', 'CNY-3', 'CNY-6', 'CNY-12',
       'GBP-1', 'GBP-2', 'GBP-3', 'GBP-6', 'GBP-12'])

#save resulting output file

```

```python
Turkey_UK_EXH.to_csv('TRY_GBP.csv',index = False)
```
