```python
def finddate(expression):
        
    year=expression.split('-')[0]
    month=expression.split('-')[1]
    date=int(year+month)

    return date
```

```python
Turkey_Producer_Indices=Turkey_Producer_Indices.rename(columns = {'Value':'PPI_TUR'})
```

```python
#Turkey_Producer_Indices=Turkey_Producer_Indices.rename(columns = {'PPI_TUR_VALUE':'PPI_TUR'})
```

```python
Turkey_Producer_Indices['PPI_TUR-1'] = Turkey_Producer_Indices['PPI_TUR'].shift(1)
Turkey_Producer_Indices['PPI_TUR-2'] = Turkey_Producer_Indices['PPI_TUR'].shift(2)
Turkey_Producer_Indices['PPI_TUR-3'] = Turkey_Producer_Indices['PPI_TUR'].shift(3)
Turkey_Producer_Indices['PPI_TUR-6'] = Turkey_Producer_Indices['PPI_TUR'].shift(6)
Turkey_Producer_Indices['PPI_TUR-12'] = Turkey_Producer_Indices['PPI_TUR'].shift(12)
```

```python
China_Producer_Indices=China_Producer_Indices.rename(columns = {'Value':'PPI_CHN'})
```

```python
China_Producer_Indices['PPI_CHN-1'] = China_Producer_Indices['PPI_CHN'].shift(1)
China_Producer_Indices['PPI_CHN-2'] = China_Producer_Indices['PPI_CHN'].shift(2)
China_Producer_Indices['PPI_CHN-3'] = China_Producer_Indices['PPI_CHN'].shift(3)
China_Producer_Indices['PPI_CHN-6'] = China_Producer_Indices['PPI_CHN'].shift(6)
China_Producer_Indices['PPI_CHN-12'] = China_Producer_Indices['PPI_CHN'].shift(12)
```

```python
UK_Producer_Indices=UK_Producer_Indices.rename(columns = {'Value':'PPI_GBR'})
```

```python
UK_Producer_Indices['PPI_GBR-1'] = UK_Producer_Indices['PPI_GBR'].shift(1)
UK_Producer_Indices['PPI_GBR-2'] = UK_Producer_Indices['PPI_GBR'].shift(2)
UK_Producer_Indices['PPI_GBR-3'] = UK_Producer_Indices['PPI_GBR'].shift(3)
UK_Producer_Indices['PPI_GBR-6'] = UK_Producer_Indices['PPI_GBR'].shift(6)
UK_Producer_Indices['PPI_GBR-12'] = UK_Producer_Indices['PPI_GBR'].shift(12)
```

```python
#UK_Producer_Indices
```

```python
mergedCSV = Turkey_Producer_Indices[['TIME','PPI_TUR','PPI_TUR-1','PPI_TUR-2',
                                     'PPI_TUR-3','PPI_TUR-6','PPI_TUR-12']].merge(UK_Producer_Indices[['TIME','PPI_GBR','PPI_GBR-1','PPI_GBR-2',
                                    'PPI_GBR-3','PPI_GBR-6','PPI_GBR-12']], 
            on = 'TIME',how = 'left')

mergedCSV=mergedCSV.rename(columns = {'TIME':'Date'})

for num in range(0,mergedCSV.shape[0]):
    mergedCSV.iloc[num,0]=finddate(mergedCSV.iloc[num,0])
mergedCSV=mergedCSV.dropna()
#save resulting output file

mergedCSV.to_csv('PPI_TUR_UK.csv',index = False)
```
