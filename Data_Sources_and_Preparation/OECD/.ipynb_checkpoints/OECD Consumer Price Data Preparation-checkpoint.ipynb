```python
import numpy as np
import pandas as pd
```

```python
df0 = pd.read_csv('OECD_Data\CPI_Indices.csv')
```

```python
df0
```

```python
def finddate(expression):
        
    year=expression.split('-')[0]
    month=expression.split('-')[1]
    date=int(year+month)

    return date
```

```python
df=df0
```

```python
df.drop('Subject',axis=1,inplace=True)
df.drop('Country',axis=1,inplace=True)
df.drop('MEASURE',axis=1,inplace=True)
df.drop('Unit Code',axis=1,inplace=True)
df.drop('Unit',axis=1,inplace=True)
df.drop('PowerCode Code',axis=1,inplace=True)
df.drop('PowerCode',axis=1,inplace=True)
df.drop('Reference Period Code',axis=1,inplace=True)
df.drop('FREQUENCY',axis=1,inplace=True)
df.drop('Frequency',axis=1,inplace=True)
df.drop('Reference Period',axis=1,inplace=True)
df.drop('Flag Codes',axis=1,inplace=True)
df.drop('Time',axis=1,inplace=True)
df.drop('Flags',axis=1,inplace=True)
```

```python
df
```

```python
Turkey_Consumer_Indices=df[(
    (df['SUBJECT']=='CPALTT01')  
   & (df['LOCATION']=='TUR') 
   & (df['Measure']=='Index 2010=100')
    )]
```

```python
China_Consumer_Indices=df[(
    (df['SUBJECT']=='CPALTT01')  
   & (df['LOCATION']=='CHN') 
   & (df['Measure']=='Index 2010=100')
    )]
```

```python
UK_Consumer_Indices=df[(
    (df['SUBJECT']=='CPALTT01')  
   & (df['LOCATION']=='GBR') 
   & (df['Measure']=='Index 2010=100')
    )]
```

```python
Turkey_Consumer_Indices=Turkey_Consumer_Indices.rename(columns = {'Value':'CPI_TUR'})
Turkey_Consumer_Indices=Turkey_Consumer_Indices.rename(columns = {'TIME':'Date'})
```

```python
Turkey_Consumer_Indices['CPI_TUR-1'] = Turkey_Consumer_Indices['CPI_TUR'].shift(1)
Turkey_Consumer_Indices['CPI_TUR-2'] = Turkey_Consumer_Indices['CPI_TUR'].shift(2)
Turkey_Consumer_Indices['CPI_TUR-3'] = Turkey_Consumer_Indices['CPI_TUR'].shift(3)
Turkey_Consumer_Indices['CPI_TUR-6'] = Turkey_Consumer_Indices['CPI_TUR'].shift(6)
Turkey_Consumer_Indices['CPI_TUR-12'] = Turkey_Consumer_Indices['CPI_TUR'].shift(12)
```

```python
for num in range(0,Turkey_Consumer_Indices.shape[0]):
    Turkey_Consumer_Indices.iloc[num,3]=finddate(Turkey_Consumer_Indices.iloc[num,3])
```

```python
China_Consumer_Indices=China_Consumer_Indices.rename(columns = {'Value':'CPI_CHN'})
China_Consumer_Indices=China_Consumer_Indices.rename(columns = {'TIME':'Date'})
```

```python
China_Consumer_Indices['CPI_CHN-1'] = China_Consumer_Indices['CPI_CHN'].shift(1)
China_Consumer_Indices['CPI_CHN-2'] = China_Consumer_Indices['CPI_CHN'].shift(2)
China_Consumer_Indices['CPI_CHN-3'] = China_Consumer_Indices['CPI_CHN'].shift(3)
China_Consumer_Indices['CPI_CHN-6'] = China_Consumer_Indices['CPI_CHN'].shift(6)
China_Consumer_Indices['CPI_CHN-12'] = China_Consumer_Indices['CPI_CHN'].shift(12)
```

```python
for num in range(0,China_Consumer_Indices.shape[0]):
    China_Consumer_Indices.iloc[num,3]=finddate(China_Consumer_Indices.iloc[num,3])
```

```python
UK_Consumer_Indices=UK_Consumer_Indices.rename(columns = {'Value':'CPI_GBR'})
UK_Consumer_Indices=UK_Consumer_Indices.rename(columns = {'TIME':'Date'})
```

```python
UK_Consumer_Indices['CPI_GBR-1'] = UK_Consumer_Indices['CPI_GBR'].shift(1)
UK_Consumer_Indices['CPI_GBR-2'] = UK_Consumer_Indices['CPI_GBR'].shift(2)
UK_Consumer_Indices['CPI_GBR-3'] = UK_Consumer_Indices['CPI_GBR'].shift(3)
UK_Consumer_Indices['CPI_GBR-6'] = UK_Consumer_Indices['CPI_GBR'].shift(6)
UK_Consumer_Indices['CPI_GBR-12'] = UK_Consumer_Indices['CPI_GBR'].shift(12)
```

```python
for num in range(0,UK_Consumer_Indices.shape[0]):
    UK_Consumer_Indices.iloc[num,3]=finddate(UK_Consumer_Indices.iloc[num,3])
```

```python
Turkey_GBR_Consumer_Indices = Turkey_Consumer_Indices[['Date','CPI_TUR','CPI_TUR-1','CPI_TUR-2',
                                     'CPI_TUR-3','CPI_TUR-6','CPI_TUR-12']].merge(UK_Consumer_Indices[['Date',
                                     'CPI_GBR','CPI_GBR-1', 'CPI_GBR-2',
                                    'CPI_GBR-3','CPI_GBR-6','CPI_GBR-12']], 
            on = 'Date',how = 'left')

# Reindex main features for the first columns 'CPI_TUR' , 'CPI_GBR'

Turkey_GBR_Consumer_Indices=Turkey_GBR_Consumer_Indices.reindex(columns=['Date', 'CPI_TUR','CPI_GBR','CPI_TUR-1', 'CPI_TUR-2', 
                                                                             'CPI_TUR-3', 'CPI_TUR-6','CPI_TUR-12', 'CPI_GBR-1', 
                                                                             'CPI_GBR-2', 'CPI_GBR-3','CPI_GBR-6', 'CPI_GBR-12'])

#save resulting output file

Turkey_GBR_Consumer_Indices.to_csv('CPI_TUR_GBR.csv',index = False)
```

```python
China_GBR_Consumer_Indices = China_Consumer_Indices[['Date','CPI_CHN','CPI_CHN-1','CPI_CHN-2',
                                     'CPI_CHN-3','CPI_CHN-6','CPI_CHN-12']].merge(UK_Consumer_Indices[['Date','CPI_GBR','CPI_GBR-1','CPI_GBR-2',
                                    'CPI_GBR-3','CPI_GBR-6','CPI_GBR-12']], 
            on = 'Date',how = 'left')


# Reindex main features for the first columns 'CPI_CHN' , 'CPI_GBR'

China_GBR_Consumer_Indices=China_GBR_Consumer_Indices.reindex(columns=['Date', 'CPI_CHN','CPI_GBR','CPI_CHN-1', 'CPI_CHN-2', 
                                                                             'CPI_CHN-3', 'CPI_CHN-6','CPI_CHN-12', 'CPI_GBR-1', 
                                                                             'CPI_GBR-2', 'CPI_GBR-3','CPI_GBR-6', 'CPI_GBR-12'])


#save resulting output file
China_GBR_Consumer_Indices.to_csv('CPI_CHN_GBR.csv',index = False)
```
