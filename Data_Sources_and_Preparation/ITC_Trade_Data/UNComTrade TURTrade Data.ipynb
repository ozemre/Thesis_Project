```python
import numpy as np
import pandas as pd
# Select Product Code 841810, 841840, or 841850
Product=841840
```

```python
#Butun Dosyalar okunur Dosya isimleri duzenlenmelidir.
Turkey_17 = pd.read_csv('Turkey Export Data\{}-2017-Turkey.txt'.format(Product),sep='\t')
Turkey_16 = pd.read_csv('Turkey Export Data\{}-2016-Turkey.txt'.format(Product),sep='\t')
Turkey_14 = pd.read_csv('Turkey Export Data\{}-2014-Turkey.txt'.format(Product),sep='\t')
Turkey_12 = pd.read_csv('Turkey Export Data\{}-2012-Turkey.txt'.format(Product),sep='\t')
Turkey_11 = pd.read_csv('Turkey Export Data\{}-2011-Turkey.txt'.format(Product),sep='\t')
Turkey_09 = pd.read_csv('Turkey Export Data\{}-2009-Turkey.txt'.format(Product),sep='\t')
Turkey_08 = pd.read_csv('Turkey Export Data\{}-2008-Turkey.txt'.format(Product),sep='\t')
Turkey_06 = pd.read_csv('Turkey Export Data\{}-2006-Turkey.txt'.format(Product),sep='\t')
```

```python
#Butun Importer bilgisine gore birlestirilir.
TUR_all = Turkey_17.merge(Turkey_16, 
                        on = 'Importers',how = 'left').merge(Turkey_14,
                        on = 'Importers',how = 'left').merge(Turkey_12,
                        on = 'Importers',how = 'left').merge(Turkey_11, 
                        on = 'Importers',how = 'left').merge(Turkey_09,
                        on = 'Importers',how = 'left').merge(Turkey_08, 
                        on = 'Importers',how = 'left').merge(Turkey_06, 
                        on = 'Importers',how = 'left')
```

```python
def finddate(expression):
        
    year=expression.split('-M')[0]
    year=year.split()[3]
    month=expression.split('-M')[1]
    month=month.split('_')[0]
    date=int(year+month)
    year=int(year)
    month=int(month)

    return date, year, month
```

```python
for num in range(0,TUR_all.shape[0]):
    TUR_all.iloc[num,7]=finddate(TUR_all.iloc[num,0])[0]
    TUR_all.iloc[num,8]=finddate(TUR_all.iloc[num,0])[1]
    TUR_all.iloc[num,9]=finddate(TUR_all.iloc[num,0])[2]
```
