```python
import numpy as np
import pandas as pd

```

```python


xls = pd.ExcelFile('Global_Trade_Data\Global Trade May 2018.xlsx')
WorldTrade = xls.parse('World', header=0,index_col=None, na_values=['NA'])
#WorldTrade = xls.parse('World', skiprows=9, header=0,index_col=None, na_values=['NA'])
```

```python
WorldTrade
```

```python
#Zaman Kaydırması Yapmak için kullanıyor. 
#Butun veri zaman ekseni açısından tam ve sıralandıktan sonra yapılması gerekiyor
WorldTrade['World-1'] = WorldTrade['World'].shift(1)
WorldTrade['World-2'] = WorldTrade['World'].shift(2)
WorldTrade['World-3'] = WorldTrade['World'].shift(3)
WorldTrade['World-6'] = WorldTrade['World'].shift(6)
WorldTrade['World-12'] = WorldTrade['World'].shift(12)
```

```python
WorldTrade
```

```python

#save resulting output file

WorldTrade.to_csv('WorldTrade.csv',index = False)
```
