```python
import numpy as np
import pandas as pd

```

```python


xls = pd.ExcelFile('Uncertainity_Data\Europe_Policy_Uncertainty_Data.xlsx')
EPU = xls.parse('European News-Based Index', header=0,index_col=None, na_values=['NA'])
#EPU Economic Politic Uncertainity 
```

```python
EPU
```

```python
EPU=EPU.dropna()
EPU.Month=EPU.Month.astype(int)
EPU.Year=EPU.Year.astype(int)
EPU['Date']= EPU['Year']*100+EPU['Month'].astype(int)
```

```python
EPU=EPU.rename(columns = {'UK_News_Index':'EPU_GBR'})
#Zaman Kaydırması Yapmak için kullanıyor. 
#Butun veri zaman ekseni açısından tam ve sıralandıktan sonra yapılması gerekiyor
EPU['EPU_GBR-1'] = EPU['EPU_GBR'] .shift(1)
EPU['EPU_GBR-2'] = EPU['EPU_GBR'].shift(2)
EPU['EPU_GBR-3'] = EPU['EPU_GBR'].shift(3)
EPU['EPU_GBR-4'] = EPU['EPU_GBR'].shift(4)
EPU['EPU_GBR-5'] = EPU['EPU_GBR'].shift(5)
EPU['EPU_GBR-6'] = EPU['EPU_GBR'] .shift(6)
EPU['EPU_GBR-7'] = EPU['EPU_GBR'].shift(7)
EPU['EPU_GBR-8'] = EPU['EPU_GBR'].shift(8)
EPU['EPU_GBR-12'] = EPU['EPU_GBR'].shift(12)
```

```python
EPU
```

```python
EPU_GBR=EPU[['EPU_GBR','Date','EPU_GBR-1','EPU_GBR-2',
                                     'EPU_GBR-3','EPU_GBR-4','EPU_GBR-5',
                                     'EPU_GBR-6','EPU_GBR-7','EPU_GBR-8','EPU_GBR-12']]

EPU_GBR=EPU_GBR.dropna()
EPU_GBR
```

```python
EPU_GBR.to_excel('EPU_GBR.xlsx',index = False)
```
