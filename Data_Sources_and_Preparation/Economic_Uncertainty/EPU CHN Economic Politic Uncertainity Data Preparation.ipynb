```python
import numpy as np
import pandas as pd

```

```python
xls = pd.ExcelFile('Uncertainity_Data\China_Policy_Uncertainty_Data.xlsx')
EPU = xls.parse('China EPU Index', header=0,index_col=None, na_values=['NA'])
#EPU Economic Politic Uncertainity 
```

```python
EPU
```

```python
EPU=EPU.dropna()
```

```python
EPU.month=EPU.month.astype(int)
EPU.year=EPU.year.astype(int)
```

```python
EPU['Date']= EPU['year']*100+EPU['month']
#EPU['Date']
```

```python
EPU=EPU.rename(columns = {'China News-Based EPU':'EPU_CHN'})
#Zaman Kaydırması Yapmak için kullanıyor. 
#Butun veri zaman ekseni açısından tam ve sıralandıktan sonra yapılması gerekiyor
EPU['EPU_CHN-1'] = EPU['EPU_CHN'] .shift(1)
EPU['EPU_CHN-2'] = EPU['EPU_CHN'].shift(2)
EPU['EPU_CHN-3'] = EPU['EPU_CHN'].shift(3)
EPU['EPU_CHN-4'] = EPU['EPU_CHN'].shift(4)
EPU['EPU_CHN-5'] = EPU['EPU_CHN'].shift(5)
EPU['EPU_CHN-6'] = EPU['EPU_CHN'].shift(6)
EPU['EPU_CHN-7'] = EPU['EPU_CHN'].shift(7)
EPU['EPU_CHN-8'] = EPU['EPU_CHN'].shift(8)
EPU['EPU_CHN-12'] = EPU['EPU_CHN'].shift(12)
```

```python
EPU
```

```python
EPU_CHN=EPU[['EPU_CHN','Date','EPU_CHN-1','EPU_CHN-2',
                                     'EPU_CHN-3','EPU_CHN-4','EPU_CHN-5',
                                     'EPU_CHN-6','EPU_CHN-7','EPU_CHN-8','EPU_CHN-12']]

#EPU_CHN=EPU_CHN.dropna()
EPU_CHN
```

```python
EPU_CHN.to_excel('EPU_CHN.xlsx',index = False)
```
