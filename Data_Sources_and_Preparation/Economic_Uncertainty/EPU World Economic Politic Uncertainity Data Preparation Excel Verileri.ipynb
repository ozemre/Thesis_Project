```python
import numpy as np
import pandas as pd

```

```python


xls = pd.ExcelFile('Global_Policy_Uncertainty_Data.xlsx')
EPU = xls.parse('Sheet1', header=0,index_col=None, na_values=['NA'])
#EPU Economic Politic Uncertainity 
```

```python
EPU
```

```python
EPU=EPU.reset_index()
EPU=EPU.dropna()
```

```python
EPU
```

```python
EPU['Date']= (EPU['level_0'].astype(int))*100+EPU['level_1'].astype(int)
#EPU['level_0']*2
#EPU['level_1']
EPU
```

```python
EPU=EPU.rename(columns = {'Global EPU Index with Current Price GDP Weights':'EPU_World'})
#Zaman Kaydırması Yapmak için kullanıyor. 
#Butun veri zaman ekseni açısından tam ve sıralandıktan sonra yapılması gerekiyor
EPU['EPU_World-1'] = EPU['EPU_World'] .shift(1)
EPU['EPU_World-2'] = EPU['EPU_World'].shift(2)
EPU['EPU_World-3'] = EPU['EPU_World'].shift(3)
EPU['EPU_World-6'] = EPU['EPU_World'].shift(6)
EPU['EPU_World-12'] = EPU['EPU_World'].shift(12)
```

```python
EPU
```

```python
EPU_World=EPU[['Date','EPU_World','EPU_World-1','EPU_World-2',
                                     'EPU_World-3','EPU_World-6','EPU_World-12']]

EPU_World=EPU_World.dropna()
EPU_World
```

```python
EPU_World.to_excel('EPU_World.xlsx',index = False)
```
