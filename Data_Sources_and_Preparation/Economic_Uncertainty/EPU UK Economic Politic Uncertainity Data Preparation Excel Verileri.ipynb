```python
import numpy as np
import pandas as pd

```

```python


xls = pd.ExcelFile('Europe_Policy_Uncertainty_Data.xlsx')
EPU = xls.parse('European News-Based Index', header=0,index_col=None, na_values=['NA'])
#EPU Economic Politic Uncertainity 
```

```python
EPU
```

```python
EPU['Date']= EPU['Year']*100+EPU['Month'].astype(int)
```

```python
EPU=EPU.rename(columns = {'UK_News_Index':'EPU_UK'})
#Zaman Kaydırması Yapmak için kullanıyor. 
#Butun veri zaman ekseni açısından tam ve sıralandıktan sonra yapılması gerekiyor
EPU['EPU_UK-1'] = EPU['EPU_UK'] .shift(1)
EPU['EPU_UK-2'] = EPU['EPU_UK'].shift(2)
EPU['EPU_UK-3'] = EPU['EPU_UK'].shift(3)
EPU['EPU_UK-6'] = EPU['EPU_UK'].shift(6)
EPU['EPU_UK-12'] = EPU['EPU_UK'].shift(12)
```

```python
EPU
```

```python
EPU_UK=EPU[['Date','EPU_UK','EPU_UK-1','EPU_UK-2',
                                     'EPU_UK-3','EPU_UK-6','EPU_UK-12']]

EPU_UK=EPU_UK.dropna()
EPU_UK
```

```python
EPU_UK.to_excel('EPU_UK.xlsx',index = False)
```
